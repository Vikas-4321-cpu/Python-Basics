2. What is an interpreter in Python?

3. What are pre-defined keywords in Python?

Examples of keywords include if, else, while, for, def, class, import, try, except, return, True, False, None, etc.
Keywords are case-sensitive and cannot be redefined or overridden within a Python program.

4. Can keywords be used as variable names?
   >>No, keywords cannot be used as variable names in Python.

   Why?
   Keywords are reserved words that have special meanings in Python. Using them as variable names would cause syntax errors because Python wouldn't know whether you're referring to a variable or a built-in function/statement.

5. What is mutability in Python?
   >>n Python, objects can be either changeable (modifiable) or unchangeable (unchangeable).
   Mutable objects, such as lists and dictionaries, can be modified after creation, while immutable objects, such as tuples and strings, cannot.
   Mutability affects how objects are stored and manipulated in memory, impacting performance, memory usage, and concurrency in Python programs.

6. Why are lists mutable, but tuples are immutable?
   >>Mutability & Immutability Explained
   Mutable objects can be changed after creation.
   Immutable objects cannot be changed once created.
   Lists (Mutable)
   Lists allow adding, removing, or modifying elements.
   They are stored as references to memory locations, meaning modifications affect the original object.
   Tuples (Immutable)
   Once a tuple is created, its elements cannot be modified.
   Tuples use fixed memory allocation, making them faster and more memory-efficient.
   They are often used when data should remain constant.
   Why Is This Important?
   Lists are used when data needs to change dynamically.
   Tuples are used for fixed, read-only data, ensuring safety and faster performance.

9. What is type casting in Python?

   Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.
   Type casting is often necessary for performing arithmetic operations, data manipulation, and input/output operations in Python programs.

10.What is the difference between implicit and explicit type casting?
   >>In Python, type casting (or type conversion) is the process of converting one data type into another. There are two types:

   - Implicit Type Casting (Automatic Conversion)
   Python automatically converts a smaller data type to a larger one without user intervention.
   No data loss occurs.
   Here, Python implicitly converts int to float to avoid data loss.

   - Explicit Type Casting (Manual Conversion)
   Done manually using functions like int(), float(), str(), etc.
   May cause data loss if converting between incompatible types.

11.What is the purpose of conditional statements in Python?
   >>Conditionals in Python are used to execute code based on the evaluation of one or more conditions.
   Python supports conditional statements such as if, elif (else if), and else to control the flow of execution.

12.How does the elif statement work?
   >> The elif statement in Python is short for "else if" and is used to test multiple conditions in sequence after an if statement. It allows you to check additional conditions if the previous if or elif conditions were not met. If none of the conditions are True, the else block is executed.

   How it works:
   - First, the if statement is checked. If it’s True, the block of code inside the if is executed, and the rest of the conditions are ignored.
   - Next, if the if statement is False, Python will check the elif (else if) condition. If it’s True, the block of code under the elif is executed.
   - Finally, if neither the if nor the elif conditions are True, the else block (if provided) will run.

13.What is the difference between for and while loops?
   >>The for and while loops in Python are both used for repeated execution of code, but they differ in how they control the iteration.

   - for Loop:
   Used for iterating over a sequence (like a list, tuple, string, or range).
   The number of iterations is determined beforehand, and the loop runs for each item in the sequence.
   Commonly used when you know how many times you want to execute the loop.
   - while Loop:
   Used for repeating a block of code as long as a certain condition is True.
   The number of iterations is not predetermined. It continues until the condition evaluates to False.
   Commonly used when you don't know how many times you need to iterate but you know the condition under which to stop.

14.Describe a scenario where a while loop is more suitable than a for loop.?
   >>A while loop is more suitable than a for loop in situations where the number of iterations is not known beforehand and is based on a condition that might change dynamically during the execution of the loop.

   Scenario Example: User Authentication
   Imagine a scenario where you want to repeatedly ask a user for their password until they enter the correct one. You don't know how many attempts it will take, so a while loop is perfect here because it continues asking the user for the password until the correct one is entered.
   Why the while loop is suitable here:
   The number of iterations is unknown because the user might enter the correct password at any point.
   The loop continues based on a condition (whether the user entered the correct password or reached the maximum attempts).
   If you used a for loop, you would need a predetermined number of attempts, but this scenario requires flexibility in the number of iterations

```python
# 8.Write a program to demonstrate the use of logical operators
# Defining some variables for the demonstration
age = 25
has_license = True
is_student = False
has_ticket = True

# Using the logical AND operator
if age >= 18 and has_license:
    print("You are eligible to drive.")  # This will execute if both conditions are True
else:
    print("You are not eligible to drive.")

# Using the logical OR operator
if has_ticket or is_student:
    print("You can enter the event.")  # This will execute if at least one condition is True
else:
    print("You cannot enter the event.")

# Using the logical NOT operator
if not is_student:
    print("You are not a student.")  # This will execute since is_student is False
else:
    print("You are a student.")

# Combining all logical operators together
if age >= 18 and (has_license or has_ticket):
    print("You are allowed to drive or enter the event.")  # This checks both conditions with AND and OR
else:
    print("You are not allowed to drive or enter the event.")
```

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types
# Taking user input
user_input = input("Enter a value: ")

# Converting the input to integer
try:
    int_value = int(user_input)
    print(f"Converted to integer: {int_value}")
except ValueError:
    print("Cannot convert to integer.")

# Converting the input to float
try:
    float_value = float(user_input)
    print(f"Converted to float: {float_value}")
except ValueError:
    print("Cannot convert to float.")

# Converting the input to boolean
# In Python, the string "True" and "False" are treated as boolean values.
# Any non-empty string is considered True, and an empty string is False.
bool_value = bool(user_input)
print(f"Converted to boolean: {bool_value}")

```

```python
# 10.Write code to demonstrate type casting with list elements
# List with string elements representing different types
my_list = ['10', '20.5', 'True', '30', '40.99']

# Converting all elements to integers where possible (ignoring non-numeric values)
int_list = []
for item in my_list:
    try:
        int_list.append(int(float(item)))  # First convert to float to handle decimals, then to int
    except ValueError:
        int_list.append(None)  # If conversion fails, append None
print("List with elements as integers:", int_list)

# Converting all elements to floats
float_list = []
for item in my_list:
    try:
        float_list.append(float(item))  # Convert all to float
    except ValueError:
        float_list.append(None)  # If conversion fails, append None
print("List with elements as floats:", float_list)

# Converting all elements to booleans
bool_list = [bool(item) for item in my_list]  # Any non-empty string will be True
print("List with elements as booleans:", bool_list)

```

```python
# 13.Write a Python program to find the sum of all even numbers between 1 and 50
# Function to calculate the sum of even numbers between 1 and 50
def sum_of_even_numbers():
    total_sum = 0  # Variable to store the sum of even numbers

    # Loop through numbers between 1 and 50 (inclusive)
    for num in range(1, 51):
        if num % 2 == 0:  # Check if the number is even
            total_sum += num  # Add the even number to the total sum

    return total_sum

# Call the function and display the result
result = sum_of_even_numbers()
print("The sum of all even numbers between 1 and 50 is:", result)
```

```python
# 14.Write a program to reverse a string using a while loop
# Function to reverse a string using a while loop
def reverse_string(input_string):
    reversed_string = ""
    index = len(input_string) - 1  # Start from the last character

    while index >= 0:  # Continue until the first character is reached
        reversed_string += input_string[index]  # Add the current character to the reversed string
        index -= 1  # Move to the previous character

    return reversed_string

# Input string from the user
user_input = input("Enter a string to reverse: ")

# Reverse the string
reversed_str = reverse_string(user_input)

# Print the reversed string
print("Reversed string:", reversed_str)
```

```python
# 15.Write a Python program to calculate the factorial of a number provided by the user using a while loop
# Function to calculate factorial using while loop
def factorial(n):
    result = 1  # Initializing result to 1 (since factorial of 0 is 1)

    while n > 0:
        result *= n  # Multiply result by the current number
        n -= 1  # Decrease n by 1
    return result

# Input from the user
try:
    num = int(input("Enter a number to calculate its factorial: "))

    # Check if the input number is negative
    if num < 0:
        print("Factorial is not defined for negative numbers.")
    else:
        # Calculate the factorial
        fact = factorial(num)
        print(f"The factorial of {num} is {fact}")
except ValueError:
    print("Please enter a valid integer.")
```
